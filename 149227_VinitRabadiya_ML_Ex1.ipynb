```python
#imports
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
import numpy as np
```

### 1. Political Party - Pie chart

```python
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/ML/kenyaElection2017.csv')

df_sorted = df.sort_values(by='%', ascending=False) #sort by percentage descending order
top2 = df_sorted.head(2) #keeping top 2 parties as they have most percentage value

#joining the other parties and percentage to be Other
others = df_sorted.iloc[2:]
other_sum = others['%'].sum()

#creating new dataframe for plotting
df_plot = pd.concat([
    top2,
    pd.DataFrame({'Party': ['Other'], '%': [other_sum]})
])

plt.figure(figsize=(10,7))
plt.pie(df_plot['%'],
        labels=df_plot['Party'],
        autopct='%1.1f%%')
plt.title('Political Parties and Vote Percentage')
plt.axis('equal')
plt.show()
#other legends
```

### 2. First 10 Fibonacci numbers

```python
fib = [0,1] #initializing first 2
for i in range(2,10):
    # fib[-1]  last number in the list
    # fib[-2]  second last number
    fib.append(fib[-1] + fib[-2])
x = list(range(10))
plt.plot(x,fib, label="Line")
list_sizes = [30, 60, 90, 120, 150, 180, 210, 240, 270, 300]
plt.scatter(x,fib, label="Dots", s=list_sizes, marker="s")
plt.title("First 10 Fibonacci numbers")
plt.xlabel("n") #no of fibonacci seq
plt.ylabel("Fibonnaci number") #fibonacci number
plt.legend()
plt.grid()
plt.show()
```

### 8 skewness

```python
np.random.seed(10) #10 is seed value,generator starts at a specific spot
scores = np.random.normal(75, 10, 100)
scores2 = np.random.normal(50, 1, 90) #creates 100data points normal distribution, 75 is mean, 10 is std deviation(how spread the scores are)
# Add some extreme scores (outliers)
outliers = np.array([20, 25, 100, 105])
data = np.concatenate([scores, outliers, scores2])
data2 = np.concatenate([scores, outliers, scores2 + 5])
 #joins the normal scores and the outliers into one single list of data

plt.boxplot([data, data2])


plt.title('Student Scores')
plt.ylabel('Score Value')
plt.show()
```



*   Position of the median line
    *   If the median is roughly in the center then distribution is symmetric
    *   If the median is closer to the bottom, Q1 then right, positive skew
    *   If the median is closer to the top, Q3 then  left,  negative skew

* Length of the whiskers
    * Longer upper whisker is right skew
    * Longer lower whisker is left skew

* Outliers
    * Interquartile Range (IQR) IQR=Q3−Q1
    * Lower bound = Q1−1.5×IQR
    * Upper bound = Q3+1.5×IQR
    * Below the lower bound is lower outlier
    * Above the upper bound is upper outlier
