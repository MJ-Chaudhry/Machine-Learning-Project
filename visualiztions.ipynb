```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("finaccess2024_datasprint.xlsx")

df.head()
```

```python
pivot_counts = df.groupby(['Sex', 'financial_status']).size().unstack(level="Sex")

pivot_counts.plot(kind="bar")
plt.title("Distribution of Sex by Financial Status")
plt.xlabel("Financial Status")
plt.ylabel("Count")
plt.xticks(rotation=0)
```

```python
financial_status = df.groupby('financial_status').size()
financial_status.plot(
    kind="pie", 
    autopct='%1.1f%%',
    title="Distribution of Financial Status")
```
